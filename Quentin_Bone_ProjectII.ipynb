##### Step 1: 
Firstly, using df.info(), shows the columns in the dataset, as well as the non-null count and the datatype of the dataset, giving an overall description of the dataset.
##### Step 2: 
Using pd.set_option('display.max_columns/rows', None) allows for all the columns/rows to be shown when using df.head/tail() and finding and counting null values. The None at the end ensures that there is no limit to the columns/rows regardless of the number. 
##### Step 3: 
Secondly, using df.head() and df.tail() gives a small insight into the dataset, with df.head() showing the first few rows and df.tail() showing the last few rows. 
##### Step 4: 
The next step is finding the null values. To find these, using df.isnull().sum() goes through the dataset and finds all the null values, sums them up and organizes them by the columns which they are from. After that, to see the null values in a more organized and concise manner, using null_counts. The '[null_counts > 0]' tells python to only count the columns with more than 0 null values. The sort_values(ascending=True) then sorts the values and lists them in ascending order. 
##### Step 5: 
Finally, in order to determine which missing values should be filled in and cleaned, calculating the percentage of null values is important. If the percentage is around 20%, then the data should be filled/cleaned. By using the same df.isnull().sum() and adding /len(df) * 100 calculates the percentage of missing values. The same null_counts > 0 is used to only include the necessary info.

##### .value_counts()
Using value_counts(), it is possible to see the most common values for certain categories. In this case, environment score and average student attendance are used to give an idea of which ranges are most common. Since the dataset is so large, using .head(25) ensures that the data is readable and the most common and relevant results are displayed. 

```python
df_new.groupby('Elementary, Middle, or High School')['School ID'].count()
```

```python
df_new.groupby('Elementary, Middle, or High School')['Environment Score'].mean()
```

```python
df_new.groupby('Elementary, Middle, or High School')['Safety Score'].mean()
```

```python
df_new.groupby('Elementary, Middle, or High School')['Average Student Attendance'].mean()
```

##### .groupby()
Using .groupby() groups a column of data by another column. For example, with the first .groupby() function, the School ID's are grouped by whether they are elementary, middle or high schools. This is repeated for the other categories like safety score, average student attendance and environment score. Since the 'Elementary, Middle, or High School' categories has three categorical values, it can be the column that is grouped by. 
##### Interesting findings:
Middle schools had a significantly lower environment score mean than high schools and elementary schools, while the safety score mean was very close for all three. 
The high school average student attendance mean was much lower compared to the elementary and middle school average student attendance mean.
